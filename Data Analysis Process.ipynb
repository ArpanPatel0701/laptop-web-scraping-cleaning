#### **URL**

##### https://www.flipkart.com/search?q=laptop&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page=1

```python
all_data = []



name = []
image_url = []
rating_star = []
ratings = []
reviews = []
processor = []
ram = []
operating_system = []
rom = []
display = []
warranty = []
actual_price = []
discount_price = []
discount = []






for j in range(1, 36):
    headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/115.0.0.0 Safari/537.36",
    "Accept-Language": "en-US,en;q=0.9",
    }

    web_page = requests.get("https://www.flipkart.com/search?q=laptop&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page={}".format(j), headers= headers).text
    soup = BeautifulSoup(web_page, "lxml")
    soup.prettify()

    container = soup.find_all('div', class_ = "tUxRFH")


    for i in container:
        row = {}

        # Name
        try:
            row['name'] = i.find('div', class_ = 'KzDlHZ').text.strip()
        except:
            row['name'] = None

        # Image URL
        try:
            row['image_url'] = i.find('img', class_ = 'DByuf4').get('src')
        except:
            row['image_url'] = None

        # Rating Star
        try:
            row['rating_star'] = i.find('div', class_ = 'XQDdHH').text.strip()
        except:
            row['rating_star'] = None

        # Ratings & Reviews
        try:
            rating_review_text = i.find('span', class_='Wphh3N').text.strip()
            parts = rating_review_text.split("&")
            row['ratings'] = parts[0].strip() if len(parts) > 0 else None
            row['reviews'] = parts[1].strip() if len(parts) > 1 else None
        except:
            row['ratings'] = None
            row['reviews'] = None

        # Processor, Ram, Operating System, Rom, Display, Warranty
        try:
            specs = i.find_all('li', class_='J+igdf')
            processor_text = ram_text = rom_text = os_text = display_text = warranty_text = None
            
            if len(specs) >= 5:
                row['processor'] = specs[0].text.strip()
                row['ram'] = specs[1].text.strip()
                row['operating_system'] = specs[2].text.strip()
                row['rom'] = specs[3].text.strip()
                row['display'] = specs[-2].text.strip()
                row['warranty'] = specs[-1].text.strip()

        
        except Exception as e:
            row['processor'] = None
            row['ram'] = None
            row['operating_system'] = None
            row['rom'] = None
            row['display'] = None
            row['warranty'] = None


        # Actual Price, Discount Price, Discount
        try:
            row['actual_price'] = i.find('div', class_ = 'yRaY8j ZYYwLA').text.strip()
        except:
            row['actual_price'] = None
            
        try:
            row['discount_price'] = i.find('div', class_='Nx9bqj _4b5DiR').text.strip()
        except:
            row['discount_price'] = None
        
        try:
            row['discount'] = i.find('div', class_='UkUFwK').text.strip()
        except:
            row['discount'] = None

            

        all_data.append(row)




# Create final DataFrame
df = pd.DataFrame(all_data)
```

```python
def extract_rom(text):
    match = re.search(r'(\d+\.?\d*\s*(GB|TB)\s*(SSD|HDD|eMMC)?)', str(text), flags=re.IGNORECASE)
    return match.group(0) if match else None

# Apply only where rom is missing
df['rom'] = df.apply(
    lambda row: extract_rom(row['name']) if pd.isna(row['rom']) or row['rom'] == "" else row['rom'],
    axis=1
)

# Optional: see only the updated rows
updated_rom_rows = df[df['rom'].notna()]

# Save or display results
```
